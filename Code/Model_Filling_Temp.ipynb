```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../Dataset/weatherAUS.csv")
```

```python

df = df[['MinTemp', 'MaxTemp', 'Rainfall', 'Sunshine']]
```

```python
df.isna().sum()
```

```python
df_Min_temp_fill = df[df['MinTemp'].isna()]
```

```python
df_Min_temp_fill
```

```python
df_Min_temp_fill.isna().sum()
```

```python
# Taking only those rows from df_Min_temp_fill where MaxTemp and Rainfall is not null dropping Sunshine

df_Min_temp_fill = df_Min_temp_fill[df_Min_temp_fill['MaxTemp'].notna() & df_Min_temp_fill['Rainfall'].notna()]
df_Min_temp_fill = df_Min_temp_fill.drop(['Sunshine'], axis=1)
df_Min_temp_fill
```

```python
# Extracting rows in which MinTemp, MaxTemp and Rainfall is not null from df
df_Min_temp = df[df['MinTemp'].notna() & df['MaxTemp'].notna() & df['Rainfall'].notna()]
df_Min_temp = df_Min_temp.drop(['Sunshine'], axis=1)
df_Min_temp
```

```python
# Using Theil-Sen Estimator to fill the missing values

from sklearn.linear_model import TheilSenRegressor
```

```python
X_Min_temp = df_Min_temp[['MaxTemp', 'Rainfall']]
y_Min_temp = df_Min_temp['MinTemp']

X_Min_temp_fill = df_Min_temp_fill[['MaxTemp', 'Rainfall']]
y_Min_temp_fill = df_Min_temp_fill['MinTemp']

regressor_Min_temp = TheilSenRegressor(random_state=0).fit(X_Min_temp, y_Min_temp)

y_Min_temp_pred = regressor_Min_temp.predict(X_Min_temp_fill)
```

```python
# Filling the missing values in df_Min_temp_fill with predicted values

df_Min_temp_fill['MinTemp'] = y_Min_temp_pred
```

```python
# Filling the NaN values in df with the predicted MinTemp values in df_Min_temp_fill at the same index

df.loc[df_Min_temp_fill.index, 'MinTemp'] = y_Min_temp_pred
df
```

```python
df.isna().sum()
```
