```python
import pandas as pd
import numpy as np
```

```python
df.drop(['Date', 'Location', 'Evaporation'], axis=1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df['RainTomorrow'].value_counts()
```

```python
from sklearn.utils import resample
```

```python
# Separate majority and minority classes

df_majority = df[df.RainTomorrow=='No']
df_minority = df[df.RainTomorrow=='Yes']
```

```python
# Upsample minority class

df_minority_upsampled = resample(df_minority, replace=True, n_samples=50000, random_state=123)
```

```python
# Combine majority class with upsampled minority class

df_upsampled = pd.concat([df_majority, df_minority_upsampled])
```

```python
df_upsampled.RainTomorrow.value_counts()
```

```python
# Encoding categorical data of RainToday and RainTomorrow

rain = {'No': 0, 'Yes': 1}

df_upsampled['RainToday'] = df_upsampled['RainToday'].map(rain)
df_upsampled['RainTomorrow'] = df_upsampled['RainTomorrow'].map(rain)
```

```python
# Encoding categorical data of WindGustDir, WindDir9am, and WindDir3pm

dir = {'N': 0, 'NNE': 1, 'NE': 2, 'ENE': 3, 'E': 4, 'ESE': 5, 'SE': 6, 'SSE': 7, 'S': 8, 'SSW': 9, 'SW': 10, 'WSW': 11, 'W': 12, 'WNW': 13, 'NW': 14, 'NNW': 15}

df_upsampled['WindGustDir'] = df_upsampled['WindGustDir'].map(dir)
df_upsampled['WindDir9am'] = df_upsampled['WindDir9am'].map(dir)
df_upsampled['WindDir3pm'] = df_upsampled['WindDir3pm'].map(dir)
```

```python
df_upsampled
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plotting a heatmap of the correlation between features ignoring the diagonal values and the upper triangle values

plt.figure(figsize=(15, 15))
sns.heatmap(df_upsampled.corr(), annot=True, cmap='coolwarm', square=True, mask=np.triu(df_upsampled.corr(), k=0))
```

```python
# Creating a Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
```

```python
X = df_upsampled.drop('RainTomorrow', axis=1)
y = df_upsampled['RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Splitting the dataset into the Training set and Test set

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler
```

```python
sclaer = StandardScaler()

X_train = sclaer.fit_transform(X_train)
X_test = sclaer.transform(X_test)
```

```python
# Training the Random Forest Classification model on the Training set

classifier = RandomForestClassifier(random_state=123)
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results

y_pred = classifier.predict(X_test)
```

```python
# Making the Confusion Matrix

from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Confusion Matrix

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# Accuracy Score

accuracy_score(y_test, y_pred)
```
