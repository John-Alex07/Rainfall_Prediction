```python
df = df.dropna()
```

```python
# Drop columns with more than 30% missing values

df = df.dropna(thresh=df.shape[0]*0.7,axis=1)
df
```

```python
df.isna().sum()
```

```python
df
```

```python
encode_Dir = {'N':0,'NNE':1,'NE':2,'ENE':3,'E':4,'ESE':5,'SE':6,'SSE':7,'S':8,'SSW':9,'SW':10,'WSW':11,'W':12,'WNW':13,'NW':14,'NNW':15}

df['WindGustDir'] = df['WindGustDir'].map(encode_Dir)
df['WindDir9am'] = df['WindDir9am'].map(encode_Dir)
df['WindDir3pm'] = df['WindDir3pm'].map(encode_Dir)
```

```python
df
```

```python
df.drop(['Date','Location'],axis=1, inplace=True)
```

```python
encoded_rain = {'No':0,'Yes':1}

df['RainToday'] = df['RainToday'].map(encoded_rain)
df['RainTomorrow'] = df['RainTomorrow'].map(encoded_rain)
```

```python
df
```

```python
df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Plotting a Count Plot with count of each category of RainTomorrow

sns.countplot(x='RainTomorrow',data=df)
```

```python
# Balancing the dataset

from sklearn.utils import resample
```

```python
df_majority = df[df.RainTomorrow==0]
df_minority = df[df.RainTomorrow==1]

df_minority_upsampled = resample(df_minority, replace=True, n_samples=40000, random_state=123)
```

```python
# Combine majority class with upsampled minority class

df_upsampled = pd.concat([df_majority, df_minority_upsampled])

# Display new class counts

df_upsampled.RainTomorrow.value_counts()
```

```python
# Plotting a Count Plot with count of each category of RainTomorrow

sns.countplot(x='RainTomorrow',data=df_upsampled)
```

```python
# Plotting the correlation matrix with clear visualization and values in an enlarged format and understandable image

plt.figure(figsize=(20,20))
sns.heatmap(df_upsampled.corr(),annot=True,cmap='coolwarm')
```

```python
# Creating a simple classification model using Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('RainTomorrow',axis=1)
y = df['RainTomorrow']

X_upsampled = df_upsampled.drop('RainTomorrow',axis=1)
Y_upsampled = df_upsampled['RainTomorrow']

X_train_upsampled,X_test_upsampled,y_train_upsampled,y_test_upsampled = train_test_split(X_upsampled,Y_upsampled,test_size=0.3,random_state=42)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)

model = LogisticRegression()
model_upsampled = LogisticRegression()

model_upsampled.fit(X_train_upsampled,y_train_upsampled)
model.fit(X_train,y_train)
```

```python
model_upsampled.score(X_test_upsampled,y_test_upsampled)
```

```python
# Exploring the data using Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf = RandomForestClassifier()
model_rf_upsampled = RandomForestClassifier()

model_rf.fit(X_train,y_train)
model_rf_upsampled.fit(X_train_upsampled,y_train_upsampled)
```

```python
model_rf_upsampled.score(X_test_upsampled,y_test_upsampled)
```

```python
model_rf.score(X_test,y_test)
```

```python
# XGBoost Classifier

from xgboost import XGBClassifier
```

```python

model_xgb_upsampled = XGBClassifier()
model_xgb = XGBClassifier()

model_xgb.fit(X_train,y_train)
model_xgb_upsampled.fit(X_train_upsampled,y_train_upsampled)
```

```python
model_xgb.score(X_test,y_test)
model_xgb_upsampled.score(X_test_upsampled,y_test_upsampled)
```

```python
# SVM

from sklearn.svm import SVC

model_svm_upsampled = SVC(probability=True)
model_svm = SVC(probability=True)

model_svm.fit(X_train,y_train)
model_svm_upsampled.fit(X_train_upsampled,y_train_upsampled)
```

```python
model_svm.score(X_test,y_test)
model_svm_upsampled.score(X_test_upsampled,y_test_upsampled)
```
